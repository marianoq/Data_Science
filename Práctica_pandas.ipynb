# **Librer√≠a Pandas**

```python
# Importar Pandas
import pandas as pd
```

```python
# Crear una serie de numeros y hallar su media
numeros = pd.Series([1,2,3,5,67,35,235,62])
numeros.mean()
```

```python
# Hallar la suma
numeros.sum()
```

```python
# Crear una serie de colores diferentes
colores = pd.Series(['rojo', 'amarillo', 'verde'])
colores
```

```python
# Otra serie con tipos de autos
tipos_autos = pd.Series(['sedan','SUV','Pick Up'])
tipos_autos
```

```python
# Comninar las series de tipos de autos y colores, DATAFRAME
tabla_autos = pd.DataFrame({'Tipo de Auto':tipos_autos, 'Colores':colores})
tabla_autos
```
